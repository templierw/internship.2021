```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

CASE = 1

df = pd.read_csv(f'./case0{CASE}/case0{CASE}_01.csv', index_col=0, parse_dates=True)
df[['load', 'pv']].describe()
```

```python
fig, ax = plt.subplots(6, 2, figsize=(15,20))

for i in range(0,12):
    df[df.index.month == i + 1][['load', 'pv']][0:24*7].plot(ax=ax[int(i/2)][i%2])

plt.show();
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,10))

sns.regplot(x=df.irrad, y=df.pv, marker="+", scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ax=ax[0][0])
sns.regplot(x=df.temp, y=df.pv, marker="+", scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ax=ax[0][1])
sns.regplot(x=df.irrad, y=df['load'], marker="+", scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ax=ax[1][0])
sns.regplot(x=df.temp, y=df['load'], marker="+", scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ax=ax[1][1])

plt.show();
```
